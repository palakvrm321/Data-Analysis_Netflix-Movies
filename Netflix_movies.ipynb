```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

Load dataset

```python
df=pd.read_csv("/content/netflix_titles.csv")
```

check the shape of a dataset


```python
df.shape
```

Check the missing values

```python
df.isnull().sum()
```

Handling the Null values

```python
# Fill missing categorical values with "Unknown"
df['director']=df['director'].fillna("Unknown")
df['cast']=df['cast'].fillna("Unknown")
df['country']=df['country'].fillna("Unknown")

# Drop rows with missing 'date_added' and 'duration' (very few rows)
df = df.dropna(subset=['date_added', 'duration'])

# Fill rating with mode (most common value)
df['rating=']=df['rating'].fillna(df['rating'].mode()[0])

```

```python
df.isnull().sum()
df
```

Remove the duplicates if exist

```python
before_duplicates=df.shape[0]
```

```python
df=df.drop_duplicates()
```

```python
after_duplicates=df.shape[0]
```

```python
print(f" Duplicates removed: {before_duplicates - after_duplicates}")
```

Standardize the Text Values

```python
df['type'] = df['type'].str.strip().str.title()
df['country'] = df['country'].dropna().str.strip().str.title()
```

```python
df['country']
```

Convert into proper DateFormat

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', format='%B %d, %Y')
```

```python
df['date_added']
```

Rename Columns

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df
```

Fix DataType

```python
df['release_year']=df['release_year'].astype(int)
```
